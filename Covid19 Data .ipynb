```python
import json
import folium
import requests
import mimetypes
import http.client
import pandas as pd
from folium.plugins import HeatMap
from pandas.io.json import json_normalize
```

```python
conn =http.client.HTTPSConnection("api.covid19api.com")
payload = " "
headers = {}
conn.request("GET" , "/summary" , payload , headers)
response = conn.getresponse()
data = response.read().decode("UTF-8")
```

```python
covid_data1 = json.loads(data)
```

```python
json_normalize(covid_data1['Countries'] ,sep=',')
```

```python
df = pd.DataFrame(covid_data1['Countries'])
```

```python
df.head()
```

```python
df.columns
```

```python
covid_data2 = df.drop(columns=['CountryCode' ,'Slug' ,'Date', 'Premium'] ,axis = 1)
covid_data2
```

```python
offered_map = folium.Map(tiles="Stamen Terrain" ,min_zoom=2, width='100%', height='100%' )
offered_map
```

```python
url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
country = f'{url}/world-countries.json'
```

```python
folium.Choropleth(geo_data=country , 
                  min_zoom =2,
                  name= 'Covid-19 Analysis' ,
                  data=covid_data2 , 
                  columns=['Country' ,'TotalConfirmed'], 
                  key_on='feature.properties.name', 
                  fill_color ='OrRd' , 
                  nan_fill_color ='black',
                  legend_name='Total Covid Confirmed Cases').add_to(offered_map)
```

```python
offered_map
```

```python
covid_data2.update(covid_data2['TotalConfirmed'].map('Total Confirmed :{}'.format))
covid_data2.update(covid_data2['TotalRecovered'].map('Total Recovered :{}'.format))
```

```python
coordinates =pd.read_csv('https://raw.githubusercontent.com/VinitaSilaparasetty/covid-map/master/country-coordinates-world.csv')

coordinates.head()
```

```python
covid_final = pd.merge(covid_data2 , coordinates ,on='Country')
```

```python
def dotPlot(point):
    folium.CircleMarker(location=[point.latitude, point.longitude],
                       radius = 5,
                       weight =2,
                       popup=[point.Country , point.TotalConfirmed , point.TotalRecovered],
                       fill_color ='#00000').add_to(offered_map)
```

```python
covid_final.apply(dotPlot , axis=1)
offered_map.fit_bounds(offered_map.get_bounds())
offered_map
```

```python
map1 = folium.Map(tiles='StamenToner' , min_zoom=1 )
map1
```

```python
deaths = covid_final['TotalDeaths'].astype(float)
```

```python
lat = covid_final['latitude'].astype(float)
```

```python
long = covid_final['longitude'].astype(float)
```

```python
map1.add_child(HeatMap(zip(lat , long , deaths), radius =0 ))
```
